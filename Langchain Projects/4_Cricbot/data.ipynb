```python
import pandas as pd

df = pd.read_csv("data/t20.csv")
df
```

```python
df["not_drop"] = df["Player"].apply(lambda x: "(IND" in x)
df
```

```python
df.drop(["Unnamed: 0", "Unnamed: 15"], axis=1, inplace=True)
df
```

```python
df = df[df["not_drop"]==True]
df
```

```python
df.drop(["not_drop"], axis=1, inplace=True)
df["Player"] = df["Player"].apply(lambda x: x.replace("(INDIA)", ""))
df
```

```python
df.drop("Span", axis=1, inplace=True)
df
```

```python
df.to_csv("data/ind.csv", index=False)
```
